```python
import pandas as pd
url = "https://raw.githubusercontent.com/djames44/Group_4/refs/heads/main/states_data_dataset.csv"
df = pd.read_csv(url)
df
```

```python
# Convert columns with numeric data stored as strings to numeric type
for column in df.select_dtypes(include=['object']).columns:
    try:
        df[column] = pd.to_numeric(df[column])
    except ValueError:
        print(f"Column '{column}' could not be converted to numeric. It may contain non-numeric values.")

# Calculate quantiles after conversion
Q1 = df.quantile(0.25, numeric_only=True)  # Use numeric_only=True to exclude non-numeric columns
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```

```python
# prompt: code for heat map to find corelation btween  State that have Medicaid Enrollment (2013)	Medicaid Enrollment (2016), Uninsured Rate (2010)	, Uninsured Rate (2015)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and already loaded

# Select the relevant columns for the heatmap
columns_for_heatmap = ['Medicaid Enrollment (2013)', 'Medicaid Enrollment (2016)',
                       'Uninsured Rate (2010)', 'Uninsured Rate (2015)']
correlation_data = df[columns_for_heatmap]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Medicaid Enrollment and Uninsured Rates')
plt.show()
```

```python
# Adding a column to calculate Medicaid enrollment changes
df['Medicaid Enrollment Change (2013-2016)'] = (
    df['Medicaid Enrollment (2016)'] - df['Medicaid Enrollment (2013)']
)

# Display the updated DataFrame with the new column
print(df[['State', 'Medicaid Enrollment (2013)', 'Medicaid Enrollment (2016)', 'Medicaid Enrollment Change (2013-2016)']])
```

```python
# Group states based on Medicaid expansion status
medicaid_expanded = df[df['State Medicaid Expansion (2016)'] == True]
medicaid_not_expanded = df[df['State Medicaid Expansion (2016)'] == False]

# Extract the state names
expanded_states = medicaid_expanded['State'].tolist()
not_expanded_states = medicaid_not_expanded['State'].tolist()

# Print results
print("States that expanded Medicaid:")
print(expanded_states)
print("\nNumber of states that expanded Medicaid:", len(expanded_states))

print("\nStates that did not expand Medicaid:")
print(not_expanded_states)
print("\nNumber of states that did not expand Medicaid:", len(not_expanded_states))
```

```python
# Grouping data by Medicaid Expansion status
grouped = df.groupby('State Medicaid Expansion (2016)')

# Calculating summary statistics for Medicaid enrollment changes
summary_stats = grouped['Medicaid Enrollment Change (2013-2016)'].agg(['mean', 'median', 'std'])

# Renaming the index for clarity
summary_stats.index = ['Non-Expansion States', 'Expansion States']

# Displaying the summary statistics
print(summary_stats)
```

```python
import matplotlib.pyplot as plt

# Grouping data by Medicaid Expansion status and calculating the mean
average_changes = df.groupby('State Medicaid Expansion (2016)')['Medicaid Enrollment Change (2013-2016)'].mean()

# Renaming the index for clarity
average_changes.index = ['Non-Expansion States', 'Expansion States']

# Plotting the bar chart
plt.figure(figsize=(8, 5))
average_changes.plot(kind='bar', color=['red', 'green'], alpha=0.7, edgecolor='black')

# Adding chart details
plt.title('Average Medicaid Enrollment Changes (2013-2016)', fontsize=14)
plt.ylabel('Average Enrollment Change', fontsize=12)
plt.xlabel('State Medicaid Expansion Status', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Box plot of Medicaid enrollment changes by Medicaid expansion status
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x='State Medicaid Expansion (2016)',
    y='Medicaid Enrollment Change (2013-2016)',
    palette=['red', 'green']
)

# Adding chart details
plt.title('Variability in Medicaid Enrollment Changes (2013-2016)', fontsize=14)
plt.xlabel('Medicaid Expansion Status (2016)', fontsize=12)
plt.ylabel('Medicaid Enrollment Change', fontsize=12)
plt.xticks(ticks=[0, 1], labels=['Non-Expansion States', 'Expansion States'], fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter data for Medicaid expansion and non-expansion groups
expansion_states = df[df['State Medicaid Expansion (2016)'] == True]
non_expansion_states = df[df['State Medicaid Expansion (2016)'] == False]

# Function to clean and convert percentage strings to numeric
def clean_percentage(x):
    try:
        return float(x.strip('%')) / 100  # Remove % and convert to fraction
    except AttributeError:
        return x  # Handle cases where x is not a string (e.g., NaN)

# Apply the cleaning function to the relevant columns
for col in ['Uninsured Rate (2010)', 'Uninsured Rate (2015)']:
    expansion_states[col] = expansion_states[col].apply(clean_percentage)
    non_expansion_states[col] = non_expansion_states[col].apply(clean_percentage)

# Calculate mean uninsured rate changes for each group
years = [2010, 2015]
expansion_rates = [
    expansion_states['Uninsured Rate (2010)'].mean(),
    expansion_states['Uninsured Rate (2015)'].mean()
]
non_expansion_rates = [
    non_expansion_states['Uninsured Rate (2010)'].mean(),
    non_expansion_states['Uninsured Rate (2015)'].mean()
]

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(years, expansion_rates, marker='o', label='Expansion States', color='blue')
plt.plot(years, non_expansion_rates, marker='o', label='Non-Expansion States', color='red')

# Add chart details
plt.title('Uninsured Rate Changes (2010-2015) by Medicaid Expansion Status', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Uninsured Rate (%)', fontsize=12)
plt.xticks(years)
plt.legend(title='Medicaid Expansion Status', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
