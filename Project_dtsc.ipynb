```python
import pandas as pd
url = "https://raw.githubusercontent.com/djames44/Group_4/refs/heads/main/states_data_dataset.csv"
df = pd.read_csv(url)
df
```

```python
import seaborn as sns

sns.boxplot(x=df['Uninsured Rate (2010)'])
```

```python
# Convert columns with numeric data stored as strings to numeric type
for column in df.select_dtypes(include=['object']).columns:
    try:
        df[column] = pd.to_numeric(df[column])
    except ValueError:
        print(f"Column '{column}' could not be converted to numeric. It may contain non-numeric values.")

# Calculate quantiles after conversion
Q1 = df.quantile(0.25, numeric_only=True)  # Use numeric_only=True to exclude non-numeric columns
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```

```python
# prompt: code for heat map to find corelation btween  State that have Medicaid Enrollment (2013)	Medicaid Enrollment (2016), Uninsured Rate (2010)	, Uninsured Rate (2015)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and already loaded

# Select the relevant columns for the heatmap
columns_for_heatmap = ['Medicaid Enrollment (2013)', 'Medicaid Enrollment (2016)',
                       'Uninsured Rate (2010)', 'Uninsured Rate (2015)']
correlation_data = df[columns_for_heatmap]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Medicaid Enrollment and Uninsured Rates')
plt.show()
```
